Extracts only the top results for each query

```python
import csv

def extract_cells(input_file, output_file1, output_file2):
    with open(input_file, 'r') as csv_file:
        csv_reader = csv.reader(csv_file)
        next(csv_reader)  
        for row in csv_reader:
            cell1 = row[2]
            cell2 = row[3]
            cell3 = row[5]
            cell4 = row[6]
           
            
            with open(output_file1, 'a') as file1:
                file1.write(cell1 + '\n')
                file1.write(cell3 + '\n')
                
            
            with open(output_file2, 'a') as file2:
                file2.write(cell2 + '\n')
                file2.write(cell4 + '\n')
                
import os


def process_csv_files(directory, output_file1, output_file2):
    for filename in os.listdir(directory):
        if filename.endswith(".csv"):
            filepath = os.path.join(directory, filename)
            extract_cells(filepath, output_file1, output_file2)

if __name__ == "__main__":
    directory_name = 'Final_Output'  
    output_file1 = 'file1_top2.txt'
    output_file2 = 'file2_top2.txt'
    process_csv_files(directory_name, output_file1, output_file2)    
```
